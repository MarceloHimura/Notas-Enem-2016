Lidar com dados omissos

```python
def faltantes(df, index, columns):
    for ind in index:
        for col in columns:
            df.loc[ind,col] = 0
    return df    
```

```python
index = df_dumy[df_dumy['TP_PRESENCA_MT'] != 1].index

columns = ['NU_NOTA_CN',
'NU_NOTA_CH',
'NU_NOTA_LC',
'NU_NOTA_MT',
'TP_STATUS_REDACAO',
'NU_NOTA_COMP1',
'NU_NOTA_COMP2',
'NU_NOTA_COMP3',
'NU_NOTA_COMP4',
'NU_NOTA_COMP5',
'NU_NOTA_REDACAO',]

escola = [
'TP_ENSINO',
'CO_ESCOLA',
'CO_MUNICIPIO_ESC',
'CO_UF_ESC',
'TP_DEPENDENCIA_ADM_ESC',
'TP_LOCALIZACAO_ESC',
'TP_SIT_FUNC_ESC',
]

municipio = [
'CO_MUNICIPIO_NASCIMENTO',
'CO_UF_NASCIMENTO',
]

```

```python
df_2 = faltantes(df_dumy, index, columns) 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
#ClassificaÃ§Ã£o
from sklearn.metrics import confusion_matrix, classification_report
#Regressao
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_squared_error
```

```python
X = df_2['NU_NOTA_MT']
Y = df_2.drop("NU_NOTA_MT",axis=1)
x_train, x_test, y_train, y_test = train_test_split(X,Y) 
```

```python
for i, c in enumerate(y_train.columns):
       print(y_train[c].dtypes,c)
```

```python
lr = LinearRegression(fit_intercept=True)
lr.fit(y_train, x_train)
predicted = lr.predict(y_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(x_test, predicted))  
print('Mean Squared Error:', metrics.mean_squared_error(x_test, predicted))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(x_test, predicted)))
```
