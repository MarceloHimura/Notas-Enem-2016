```python
for c in df.columns:
    if str(df[c].dtype) == 'object':
        df.drop(c, inplace=True, axis=1)
```

```python
drop = ['NU_ANO','Unnamed: 0','Q029','Q030','Q031','Q032','Q033','Q041','CO_UF_ENTIDADE_CERTIFICACAO','TP_ENSINO',
        'CO_ESCOLA','CO_MUNICIPIO_ESC','CO_UF_ESC','TP_DEPENDENCIA_ADM_ESC','TP_LOCALIZACAO_ESC','TP_SIT_FUNC_ESC']
df.drop(drop, axis=1, inplace=True)
df['TP_ESTADO_CIVIL'].fillna(0, inplace=True)
df['NU_NOTA_MT'].dropna(inplace=True,axis=0)
```

```python
df.dropna(inplace=True)
```

```python
for na in df.columns:
    if df[na].isna().sum() > 0:
        print(na, df[na].isna().sum())
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
normal = preprocessing.MinMaxScaler()
df_fit = normal.fit_transform(df.drop(['NU_NOTA_MT'],axis=1))
df_normal = pd.DataFrame(df_fit)
df_normal.columns = df.drop(['NU_NOTA_MT'],axis=1).columns
df_normal['NU_NOTA_MT'] = df['NU_NOTA_MT'].reset_index(drop=True)
```

```python
for na in df_normal.columns:
    if df_normal[na].isna().sum() > 0:
        print(na, df[na].isna().sum())
```

```python
df_normal.head()
```

```python
X = df_normal['NU_NOTA_MT']
Y = df_normal.drop('NU_NOTA_MT', axis=1)
x_train, x_test, y_train, y_test = train_test_split(X,Y)
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(y_train, x_train)
pred = model.predict(y_test)
```

```python
show = pd.DataFrame({"previsto":pred,'real':x_test})
show.head(10).reset_index(drop=True)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(x_test, pred))  
print('Mean Squared Error:', metrics.mean_squared_error(x_test, pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(x_test, pred)))
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso, Ridge, SGDRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline, FeatureUnion

X_train, X_test, Y_train, Y_test =  train_test_split(X, Y, test_size = 0.33)

pipe = Pipeline([
    ('scale', StandardScaler()),
    ('regr', Lasso())
])

param_grid = [
    {
        'regr': [Lasso(), Ridge()],
        'regr__alpha': np.logspace(-4, 1, 6),
    },
    {
        'regr': [SGDRegressor()],
        'regr__alpha': np.logspace(-5, 0, 6),
        'regr__max_iter': [500, 1000],
    },
]

grid = GridSearchCV(pipe, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid.fit(np.array(X_train).reshape(-1, 1), Y_train)

predicted = grid.predict(np.array(X_test).reshape(-1, 1))

print('Score:\t{}'.format(grid.score(np.array(X_test).reshape(-1, 1), Y_test)))
```
